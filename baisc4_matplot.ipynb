```python
import matplotlib.pylab as pyl;
import numpy as np;
```

```python
x=[1,2,3,4,8];
y = [5,7,2,1,5];
```

```python
pyl.plot(x,y);
pyl.show();
```

```python
pyl.plot(x,y,'or');
pyl.show();
```

```python
'''
c-青色
r-红色
m-品红
g-绿色
b-蓝色
y-黄色
k-黑色
w-白色
'''
'''
- 直线
-- 虚线
-. -.形式
: 细小虚线

'''
pyl.plot(x,y,'bs:')
```

```python
pyl.title("This is a title");
pyl.xlabel("age");
pyl.ylabel("temp");
pyl.xlim(0,15);#0-20的间距
pyl.ylim(0,10);
pyl.plot(x,y,'s:');
pyl.show();
```
